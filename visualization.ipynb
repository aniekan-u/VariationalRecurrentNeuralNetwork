```python
root_dir = "viz_archive/" #"saves_no_loss/"

dirs = [d for d in listdir(root_dir) if isdir(join(root_dir, d))] # different model hyperparam combinations
combinations = len(dirs)

files  = [listdir(root_dir+dirs[i]+"/") for i in range(combinations)] # [[snapshots_model_1], [snapshots_model_2],...]

pred_activities = []
device = torch.device('cuda:0')

for i, combination in enumerate(files):
    # Looping over hyperparam combinationss
    print("Combination", dirs[i])
    
    seq_len_test = int(dirs[i][dirs[i].find("sl")+2:dirs[i].find("s1_")])
    epochs = [] # filenames of model combination snapshots
    for file in files[i]:
        if file[-4:] == ".pth":
            epochs.append(int(file[str(file).find("dict_")+5:str(file).find(".pth")]))
    print(epochs)
    ord_files = np.array([files[i][k] for k in np.argsort(epochs) if files[i][k][-4:] == ".pth"])
    preds = np.zeros((seq_len_test, amount_neur, len(ord_files))) # neuron sequences for each snapshot
    for j, filename in enumerate(ord_files):
        # Looping over snapshots (ordered by epoch)
        
        x_dim = int(dirs[i][dirs[i].find("x")+1:dirs[i].find("x")+4])
        h_dim = int(dirs[i][dirs[i].find("h")+1:dirs[i].find("h")+3])
        z_dim = int(dirs[i][dirs[i].find("z")+1:dirs[i].find("la")])
        
        print(f'x: {x_dim}, h: {h_dim}, z: {z_dim}')
        print(root_dir+dirs[i]+"/"+ord_files[j])
        
        model = VRNN(x_dim, h_dim, z_dim, n_layers).to(device)
        model.load_state_dict(torch.load(root_dir+dirs[i]+"/"+ord_files[j], map_location=device))
        model.eval()
        
        train_loader = torch.utils.data.DataLoader(nwb_train, batch_size=128)
        (data, _, _) = next(iter(train_loader))
        data = data.to(torch.float)
        data = data.to(device)
        data = data.squeeze().transpose(0,1)
        data = (data - data.min()) / (data.max() - data.min())
        
        enc_means = [tens[[0],:].cpu().detach().numpy() for tens in model(data)[3][0]]
        enc_means = np.vstack(enc_means)

        plt.plot(enc_means)
        plt.show()
        
        preds[:,:,j] = enc_means[:, neur_idx]
        
    pred_activities.append(preds)
    
```

```python
def compare_plot(prediction, groundtruth):
    # Array comes in shape (seq_len, neurons, steps)    
    
    steps = int(np.shape(prediction)[2]) # epochs of snapshots
    neurons = int(np.shape(prediction)[1])
    
    def visualize(array, mode, prediction):
        if mode == 'prediction':
            for step in range(steps):
                axs[step, 0].plot(array[:,0:neurons,step])
        elif mode == 'ground_truth':
            for step in range(steps):
                axs[step, 1].plot(array[:,0:neurons])
    
    figure(figsize=(8, 10), dpi=80)
    fig, axs = plt.subplots(steps, 2)
    fig.set_size_inches(10, steps*4)
    visualize(prediction, 'prediction', steps)
    visualize(vis_data, 'ground_truth', steps)
    plt.show()
```
